```python
## applying the stopwords
import re
corpus=[]
for i in  range (len(sentences)):
    text=re.sub('[^a-zA-Z]',' ',sentences[i])
    text=text.lower()
    text=text.split()
    text=[lemmatizer.lemmatize(word) for word in text if not word in set(stopwords.words('english'))]
    text=' '.join(text)
    corpus.append(text)
```
